```python
import csv

epoch = []
train_loss = []
val_tfloss = []
kendall_tau = []
adj_coherence = []

with open('log.csv') as file:
    reader = csv.DictReader(file)
    for row in reader:
        epoch.append(row['epoch'])
        train_loss.append(row['train_loss'])
        val_tfloss.append(row['val_tfloss'])
        kendall_tau.append(row['kendall_tau'])
        adj_coherence.append(row['adj_coherence'])
```
